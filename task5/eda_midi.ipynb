```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
FEATURE_CSV = "features.csv"
# Load extracted features
df = pd.read_csv(FEATURE_CSV)
df.head()
```

```python

# 1. Pitch Distribution
plt.figure(figsize=(8, 4))
plt.hist(df["note_pitch"], bins=range(21, 109), edgecolor="black")
plt.title("Pitch Distribution")
plt.xlabel("MIDI Pitch")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 2. Duration Distribution
plt.figure(figsize=(8, 4))
plt.hist(df["duration"], bins=10, edgecolor="black")
plt.title("Note Duration Distribution")
plt.xlabel("Duration (s)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 3. Velocity Distribution
plt.figure(figsize=(8, 4))
plt.hist(df["velocity"], bins=range(0, 128, 5), edgecolor="black")
plt.title("Velocity Distribution")
plt.xlabel("Velocity")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

# 4. Scatter Plot: Start Time vs. Pitch (first 5 seconds)
subset = df[df["start_time"] <= 5.0]
plt.figure(figsize=(8, 4))
plt.scatter(subset["start_time"], subset["note_pitch"], s=10, alpha=0.6)
plt.title("Start Time vs. Pitch (first 5s)")
plt.xlabel("Start Time (s)")
plt.ylabel("MIDI Pitch")
plt.tight_layout()
plt.show()
```
